# Data Visualization

## Univarient

```python
fig, ((ax0, ax1)) = plt.subplots(nrows=1, ncols=2)
ax0.set_title("Histogram of Runs")
ax0.hist(df_bat.Runs);

ax1.set_title("Histogram of Strike Rate")
ax1.hist(df_bat.SR);

```

## Bivarient

```python
plt.figure(figsize=(10,5))
sns.barplot(x='SR', y='Player',hue='50', data=Top_batsmen.head(15))
plt.title('Top Batsman of IPL 2022')
plt.xlabel('Strike Rate')
plt.ylabel('Player')
plt.show()
```

```python
import matplotlib.pyplot as plt

# create a bar chart of Runs by player
plt.figure(figsize=(10,5))
plt.bar(Top_batsmen['Player'][:5], Top_batsmen['Runs'][:5])
plt.title('Runs by player')
plt.xlabel('Player')
plt.ylabel('Runs')

plt.show()
```

```python
import matplotlib.pyplot as plt

# sort the dataframe by Runs and select top 10 players
top10 = Top_batsmen.sort_values(by='Runs', ascending=False).head(10)

# create a horizontal bar chart of top 10 players by Runs
plt.barh(top10['Player'], top10['Runs'])
plt.title('Top 10 players by Runs')
plt.xlabel('Runs')
plt.ylabel('Player')

plt.show()
```

```python
# create the line chart of Runs vs. Inns
plt.plot(Top_batsmen['Inns'], Top_batsmen['Runs'])
plt.title('Runs vs. Inns')
plt.xlabel('Innings')
plt.ylabel('Runs')
```

```python
sns.barplot(x='SR', y='Player', hue='100', data=Top_batsmen[:15]);
plt.title('Top Batsman of IPL 2022')
plt.xlabel('Strike Rate')
plt.ylabel('Player')
plt.show()
```

```python
import seaborn as sns

# create a correlation matrix for the dataframe
corr_matrix = Top_batsmen.corr()

# plot the heatmap
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True)

plt.title('Correlation Matrix')
plt.show()
```

```python
import seaborn as sns

# select the columns of interest
subset = Top_batsmen[['Mat', 'Runs', 'SR']]

# create a correlation matrix for the subset
corr_matrix = subset.corr()

# plot the heatmap
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True)

plt.title('Correlation Matrix')
plt.show()
```

```python
import seaborn as sns

# create a pivot table with Runs as values and Inns and Player as rows/columns
table = Top_batsmen.pivot_table(values='Runs', index='Inns', columns='Player')

# plot the heatmap
sns.heatmap(table, cmap='coolwarm', annot=True, fmt='g')

plt.title('Runs by Player and Innings')
plt.show()
```
